# ELIMINACION DE DUPLICADOS

Eliminar duplicados poniendo la info de relaciones y entidades como nuevas columnas del primer registro

```python
path = "C:/Users/andre/OneDrive/Escritorio/TFM/KAG/AI-KG/AI-KG Dataset/DATA/"
file = f"{path}res1.csv"
```

```python
df = pd.read_csv(file, usecols=['text_id', 'relaciones', 'hlight_b'])
```

```python
df
```

```python
df["dup_id"] = df.groupby("text_id").cumcount() + 1
```

```python
df[df["text_id"] == "arXiv:1711.00789"]
```

```python
# Pivotear: convertir filas duplicadas en columnas
df_pivot = df.pivot(index="text_id", columns="dup_id")
```

```python
df_pivot
```

```python

# Aplanar nombres de columnas
df_pivot.columns = [
    f"{col[0]}" if col[1] == 1 else f"{col[0]}_{col[1]}"
    for col in df_pivot.columns
]

# Resetear índice para volver a tener ID como columna
df_final = df_pivot.reset_index()
```

```python
df_final
```

```python
df_final[df_final.duplicated()]
```

```python
path = "C:/Users/andre/OneDrive/Escritorio/TFM/KAG/AI-KG/AI-KG Dataset/DATA/"
out_file = f"{path}res1_group.csv"

df_final.to_csv(out_file, index=False)
```
