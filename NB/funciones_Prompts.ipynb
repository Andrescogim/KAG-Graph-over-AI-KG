# 2. CLASE

```python
class LlamadaLLM:
    def __init__(self, modelo):
        pass

    def modelos_disponibles():
        
        """Muestra los modelos disponibles"""
        
        for model in ollama.list().models:
            print(f"Modelo: {model.model} ; N_Parametros = {model.details.parameter_size}")



    def chat(modelo, prompt):
        response = ollama.chat(
            model = modelo,
            messages = [{
                'role': 'user',
                'content': prompt
                }]
        )
        return response.message.content



    def generate(modelo, prompt):

        response = ollama.generate(
            model = modelo,
            prompt = prompt
        )
        return response.response
```

```python
LlamadaLLM.modelos_disponibles()
```

```python
LlamadaLLM.chat('phi3', 'Why is the sky blue?')
```

```python
LlamadaLLM.chat('phi3', 'Why is the sky Orange?')
```

```python
LlamadaLLM.generate('phi3', 'Why is the sky Orange?')
```
