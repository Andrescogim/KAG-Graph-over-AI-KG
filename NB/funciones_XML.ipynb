```python
class LecturaXML:
    def __init__(self, ruta):
        self.tree = ET.parse(ruta)
        self.root = self.tree.getroot()

    def get_abs(self):
        
        """Extrae tuplas: (id, abstracts) del XML"""
        
        abs = []
        for triple in self.root.findall(".//triple"):
            for text in triple.findall("text"):
                text_id = text.get("id")
                abs_elem = text.find("abs")
                # print(f"text_id: {text_id}; abs: {abs_elem.text}")
                # time.sleep(1)
                if abs_elem is not None:
                    abs.append((text_id, abs_elem.text))

        return abs
```
